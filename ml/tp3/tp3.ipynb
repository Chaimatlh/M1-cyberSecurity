### EXO 1 :

### Heart Attack Analysis & Prediction Dataset


```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.tree import plot_tree

import matplotlib.pyplot as plt
```

### Charger le dataset "heart.csv"

```python
heart = pd.read_csv("heart.csv")
```

### Afficher les données

```python
print(heart.head())
```

### Proposer deux modèles à base d'arbre de décision : un basé sur le gain d'information et l'autre sur l'indice de Gini

### 1-Séparer les caractéristiques (X) de la variable cible (y)

```python
from sklearn.model_selection import train_test_split

#Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(heart.drop(" target", axis=1), heart[" target"], test_size=0.2, random_state=42)
```

### Modèle basé sur le gain d'information

```python
model_info_gain = DecisionTreeClassifier(criterion='entropy')
model_info_gain.fit(X_train, y_train)
```

### Modèle basé sur l'indice de Gini

```python
model_gini = DecisionTreeClassifier(criterion='gini')
model_gini.fit(X_train, y_train)
```

 ### Comparer les deux modèles
### Calculer l'accuracy (précision) pour chaque modèl

```python
accuracy_info_gain = model_info_gain.score(X_test, y_test)
accuracy_gini = model_gini.score(X_test, y_test)
```

 ### Afficher l'accuracy


```python
print("Précision (gain d'information) :", accuracy_info_gain)
print("Précision (indice de Gini) :", accuracy_gini)
```

### Visualiser les résultats graphiquement
### Visualisation du modèle basé sur le gain d'information
### Visualisation du modèle basé sur l'indice de Gini

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(12, 8))
plot_tree(model_info_gain, feature_names=heart.columns[:-1].tolist(), class_names=['No Disease', 'Disease'], filled=True)
plt.show()


plt.figure(figsize=(12, 8))

plot_tree(model_gini, feature_names=heart.columns[:-1].tolist(), class_names=['No Disease', 'Disease'], filled=True)

plt.show()
```


### Proposition de deux modèles CART
### Modèle 1:

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

# Load the dataset
iris_df = pd.read_csv('IRIS.csv')

# Define the variables
sepal_length = iris_df['sepal_length']
sepal_width = iris_df['sepal_width']
petal_length = iris_df['petal_length']
petal_width = iris_df['petal_width']
iris_class = iris_df['species']

# Create the CART model
model_1 = DecisionTreeClassifier(max_depth=3)

# Train the model
model_1.fit(iris_df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']], iris_class)
```

### Modèle 2:

```python
# Créer un modèle CART avec une profondeur maximale de 5
model_2 = DecisionTreeClassifier(max_depth=5)

# Entraîner le modèle
model_2.fit(iris_df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']], iris_class)
```
